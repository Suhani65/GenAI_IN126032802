```python
# Task 1: Replace hardcoded prompt using PromptTemplate

template1 = "Explain {topic} in simple terms for beginners"

prompt1 = PromptTemplate(
    input_variables=["topic"],   # required
    template=template1
)

print("Task 1 Output:")
print(prompt1.format(topic="Artificial Intelligence"))
```

```python
# Task 2: Multi-input prompt system

template2 = "Explain {topic} for {audience} in a {tone} tone"

prompt2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=template2
)

print("\nTask 2 Output:")
print(prompt2.format(topic="AI", audience="beginners", tone="friendly"))
print(prompt2.format(topic="Python", audience="kids", tone="fun"))
print(prompt2.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

```python
# Task 3: Prompt variations

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print("\nTask 3 Output:")
print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python
# Task 4: Chat-based prompt system

def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher."
        user_msg = "Explain {topic} clearly."

    elif role == "interviewer":
        system_msg = "You are an interviewer."
        user_msg = "Ask interview questions about {topic}."

    elif role == "motivator":
        system_msg = "You are a motivational speaker."
        user_msg = "Motivate someone to learn {topic}."

    else:
        system_msg = "You are a helpful assistant."
        user_msg = "Explain {topic}"

    chat_prompt = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", user_msg)
    ])

    return chat_prompt.format_messages(topic=topic)

print("\nTask 4 Output:")
messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

```python
# Task 5: Input validation

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    # Validation logic
    if audience not in valid_audience:
        print("Invalid audience → defaulting to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → defaulting to 'formal'")
        tone = "formal"

    return audience, tone

print("\nTask 5 Output:")
print(validate_inputs("kids", "cool"))  # invalid input test
```

```python
# Task 6: Prompt generator pipeline

def generate_prompt(topic, audience, tone, style):

    # Step 1: Validate input
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Select template based on style
    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"

    elif style == "interview":
        template = "Generate interview questions on {topic} for {audience} in a {tone} tone"

    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"

    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    # Step 3: Create prompt
    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    # Step 4: Generate output
    return prompt.format(topic=topic, audience=audience, tone=tone)

print("\nTask 6 Output:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# Task 7: Reusability test

reuse_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

test_inputs = [
    ("AI", "beginners"),
    ("Python", "students"),
    ("Machine Learning", "engineers"),
    ("Cloud Computing", "developers"),
    ("Cybersecurity", "analysts")
]

print("\nTask 7 Output:")
for topic, audience in test_inputs:
    print(reuse_template.format(topic=topic, audience=audience))
```
